```python
# utility to extract the video id - the text following the first "=" and before the first "&" or the end of the string.
def extract_text(s, char1="=", char2="&"):
    pattern = f"{re.escape(char1)}(.*?)(?:{re.escape(char2)}|$)"
    match = re.search(pattern, s)
    return match.group(1) if match else None


def delete_files_with_extensions(directory, extensions):
    for filename in os.listdir(directory):
        if any(filename.endswith(ext) for ext in extensions):
            os.remove(os.path.join(directory, filename))


```

```python
def chop_and_resample_wavs(directory, n_seconds, sr, output_dir, digits=3):
    """
    Reads all WAV files in a directory, chops them into n-second segments,
    resamples to a standardized sample rate (sr), selects the second channel if multi-channel,
    and saves to a specified output directory.

    Args:
        directory (str): Path to the directory containing WAV files.
        n_seconds (int): Length of each segment in seconds.
        sr (int): Target sample rate for resampling.
        output_dir (str): Path to the directory for saving chopped segments.
    """

    for filename in os.listdir(directory):
        if filename.endswith(".wav"):
            filepath = os.path.join(directory, filename)
            data, orig_sr = sf.read(filepath)

            # Select second channel if multi-channel, otherwise keep first channel
            if len(data.shape) > 1:
                data = data[:, 1]  # Select second channel by indexing
            else:
                data = data

            # Resample if necessary using soxr library
            if orig_sr != sr:
                print(f'')
                resampled_data = soxr.resample(
                    data, orig_sr, sr, quality=0  # High-quality resampling mode
                )
                data = resampled_data

            # Calculate the number of segments and avoid creating segments less than 4 seconds
            num_segments = max(0, int((data.shape[0] - (n_seconds * sr)) // (sr * n_seconds)))

            # Chop into segments and save, excluding ending portions less than 4 seconds
            for i in range(num_segments):
                start_index = i * sr * n_seconds
                end_index = min(start_index + sr * n_seconds, data.shape[0])  # Adjust end index to avoid exceeding data length

                if end_index - start_index >= sr * 4:  # Ensure segment length is at least 4 seconds
                    segment = data[start_index:end_index]

                    # Create new filename based on original name and segment index
                    output_filename = os.path.splitext(filename)[0] + f"_segment_{str(i).zfill(digits)}.wav"
                    output_filepath = os.path.join(output_dir, output_filename)

                    sf.write(output_filepath, segment, sr)

                    print(f"Processed segment {i} from '{filename}' and saved as '{output_filename}'")
```

```python

# set your parameters
n_seconds = 4  # Segment length in seconds
sr = 16000  # Target sample rate

if not os.path.exists(SEGMENTDIR ):
    os.makedirs(SEGMENTDIR, exist_ok=True )

chop_and_resample_wavs(DATADIR, n_seconds, sr,SEGMENTDIR)
```
